# PM10

```python
from sklearn.model_selection import train_test_split
```

```python
X = data_new[['so2','co','o3','no2','stasiun_DKI1 (Bunderan HI)','stasiun_DKI2 (Kelapa Gading)','stasiun_DKI3 (Jagakarsa)','stasiun_DKI4 (Lubang Buaya)','stasiun_DKI5 (Kebon Jeruk)','stasiun_DKI5 (Kebon Jeruk) Jakarta Barat']]
y = data_new['pm10']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state = 0)
```

```python
df_interpolate_lin_Pm10 = data_new['pm10'].interpolate(method='linear',axis=0)
```

```python
DataVisualPm10 = df_interpolate_lin_Pm10
DataVisualPm10.plot(title ='Raw Data Pm10', color = "red") 
plt.tight_layout()
plt.show()
```

```python
DataDlPm10 = df_interpolate_lin_Pm10
# Convert the dataframe to a numpy array
datasetdlpm10 = DataDlPm10.values
# Get the number of rows to train the model on
training_data_len_pm10 = int(np.ceil( len(datasetdlpm10) * .70))

training_data_len_pm10
```

```python
datasetdlpm10 = datasetdlpm10.reshape(-1,1)
datasetdlpm10
```

```python
# Create the training data set 
# Create the scaled training data set
train_data_pm10 = datasetdlpm10[0:int(training_data_len_pm10), :]
# Split the data into x_train and y_train data sets
x_train = []
y_train = []

for i in range(30, len(train_data_pm10)):
    x_train.append(train_data_pm10[i-30:i, 0])
    y_train.append(train_data_pm10[i, 0])
    #if i<= 100:z
        #print(x_train)
        #print(y_train)
        #print()
        
# Convert the x_train and y_train to numpy arrays 
x_train, y_train = np.array(x_train), np.array(y_train)

# Reshape the data
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
model = Sequential()
model.add(LSTM(units = 64, return_sequences=True, input_shape= (x_train.shape[1], 1)))
model.add(LSTM(units = 64, return_sequences=True))
model.add(LSTM(units = 64, return_sequences=False))
model.add(Dense(units = 1, activation="relu"))

# Compile the model
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='mean_squared_error')

history = model.fit(x_train, y_train, batch_size=100, epochs = 50)
```

```python
plt.figure(figsize=(14,6))
plt.plot(history.history["loss"], label="loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
# Create the testing data set
# Create a new array containing scaled values from index 1543 to 2002 
test_data_30_Pm10 = datasetdlpm10[training_data_len_pm10 - 30: , :]
# Create the data sets x_test and y_test
x_test_30_Pm10 = []
y_test_30_Pm10 = datasetdlpm10[training_data_len_pm10:, :]
for i in range(30, len(test_data_30_Pm10)):
    x_test_30_Pm10.append(test_data_30_Pm10[i-30:i, 0])
    
# Convert the data to a numpy array
x_test_30_Pm10 = np.array(x_test_30_Pm10)

# Reshape the data
x_test_30_Pm10 = np.reshape(x_test_30_Pm10, (x_test_30_Pm10.shape[0], x_test_30_Pm10.shape[1], 1 ))

# Get the models predicted price values 
predictions_30_Pm10 = model.predict(x_test_30_Pm10)
#predictions_30 = scaler.inverse_transform(predictions_30)
```

```python
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
mae=mean_absolute_error(y_test_30_Pm10, predictions_30_Pm10)
mse=mean_squared_error(y_test_30_Pm10, predictions_30_Pm10)
rmse=np.sqrt(mse) 
metric=[mae,mse,rmse]
evaluate=['MAE','MSE','RMSE']
performance_evaluate=pd.DataFrame([evaluate,metric])
performance_evaluate
```

```python
train_30_Pm10 = DataDlPm10[:training_data_len_pm10]
valid_30_Pm10 = df_interpolate_lin_Pm10[training_data_len_pm10:]
```

```python
train_30_Pm10 = pd.DataFrame(train_30_Pm10)
valid_30_Pm10 = pd.DataFrame(valid_30_Pm10)
valid_30_Pm10['Predictions'] = predictions_30_Pm10
valid_30_Pm10
```

```python
plt.figure(figsize=(14,6))
plt.plot(train_30_Pm10['pm10'], label="Traning",c = "g")
plt.plot(valid_30_Pm10['pm10'], label="Validation",c = "b")
plt.plot(valid_30_Pm10['Predictions'], label="predict", c = "r")
plt.legend()
plt.show()
```
